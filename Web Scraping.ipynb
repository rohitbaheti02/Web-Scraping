## Scraping data from multiple pages

We will use for loop, which changes the page number

```python
prod_name=[]
prod_price=[]
for i in range(1,11):
    # the below link will iterate number of pages
    url='https://www.flipkart.com/search?q=tv&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_1_2_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=tv&requestId=28074af0a-845f-4393-ac9d-6aac7f1ccdf0&as-backfill=on&page='+str(i)
    page=requests.get(url)
    soup=BeautifulSoup(page.text,'lxml')
    master_container=soup.find_all('div',{'class':'_2kHMtA'})
    for j in master_container:
        name=j.find('div',{'class':'_4rR01T'}).text
        prod_name.append(name)
        price=j.find('div',{'class':'_30jeq3 _1_WHN1'}).text
        prod_price.append(price)
```

```python
final=pd.DataFrame({'Product Name':prod_name,"Product Price":prod_price})
```
